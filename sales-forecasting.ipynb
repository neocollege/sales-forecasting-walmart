importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from scipy.special import boxcox1p
import seaborn as sns
```

importing the data

```python
features = pd.read_csv('features.csv')
store = pd.read_csv('stores.csv')
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train = train.groupby(['Store','Date'])['Weekly_Sales'].sum()
train = train.reset_index()
train.head(10)
```

```python
# merging train and features
data = pd.merge(train, features, on=['Store','Date'],how='inner')
data.head(10)
```

```python
# merging store and data
data = pd.merge(data,store,on=['Store'],how='inner')
data.head(10)
```

```python
# sorting the data by date
data = data.sort_values(by='Date')
data.head(10)
```

analyzing the data

```python
# count plot of type
sns.countplot(x='Type', data=data)
```

```python
# box plot of type and Weekly_Sales
sns.boxplot(x='Type',y='Weekly_Sales',data=data)
```

```python
# histogram of weekly sales
data['Weekly_Sales'].plot.hist()
```

```python
# count plot of isHoliday
sns.countplot(x='IsHoliday',data=data)
```

```python
# checking the null values of data
data.isnull().sum()
```

```python
# same, using heatmap
sns.heatmap(data.isnull(),yticklabels=False,cmap='viridis')
```

Manipulating the data

```python
data=data.drop(['MarkDown1','MarkDown2','MarkDown3','MarkDown4','MarkDown5'], axis=1)
data.head(10)
```

```python
# checking for null values
data.isnull().sum()
```

```python
# converting IsHoliday to Holiday, an integer value
data['Holiday'] = [int(i) for i in list(data['IsHoliday'])]
data.head(10)
```

```python
# converting type into integer by one-hot encoding
Type_dummy = pd.get_dummies(data['Type'],drop_first=True)
Type_dummy.head()
```

```python
data = pd.concat([data, Type_dummy], axis=1)
data.head(10)
```

```python
data = data.drop(['Type','IsHoliday'],axis=1)
data.head(10)
```

Performing Machine Learning tasks on the dataset

Splitting the data into train and test sets

```python
X = data.drop(['Weekly_Sales','Store','Date'],axis=1)
y = data['Weekly_Sales']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

Applying Linear Regression and fitting the training data into it

```python
lr = LinearRegression(normalize=True)
lr.fit(X_train, y_train)
```

Predicting data for test values

```python
y_pred = lr.predict(X_test)
plt.plot(y_test, y_pred, 'ro')
plt.plot(y_test,y_test, 'b-')
plt.show()
```

Evaluating the model by calculating errors by the root mean square error and r-squared

```python
root_mean_square_error = np.sqrt(np.mean(np.square(y_test-y_pred), axis=0))
print(root_mean_square_error)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test,y_pred)
print(r2)
```
